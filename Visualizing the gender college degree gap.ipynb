In this guided project, we'll extend the work we did in the last two missions on visualizing the gender gap across college degrees. So far, we mostly focused on the STEM degrees but now we will generate line charts to compare across all degree categories. In the last step of this guided project, we'll explore how to export the final diagram we create as an image file.

- Because there are seventeen degrees that we need to generate line charts for, we'll use a subplot grid layout of 6 rows by 3 columns.

- Disable the x-axis labels for all line charts except the bottommost line charts in each column.

- For all plots:

    Enable just the y-axis labels at 0 and 100.

- For all plots:

    Generate a horizontal line with the following properties:
        Starts at the y-axis position 50
        Set to the third color (light gray) in the Color Blind 10 palette
        Has a transparency of 0.3

- Export the figure containing all of the line charts to "gender_degrees.png".

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats=["Foreign Languages","English","Communications and Journalism","Art and Performance","Social Sciences and History"]
other_cats=["Health Professions","Public Administration","Education","Agriculture","Business","Architecture"]


fig = plt.figure(figsize=(16, 20))

for sp in range(1,7):
    ax = fig.add_subplot(6,3,sp+(sp-1)*2)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp-1]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp-1]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255,171/255,171/255),alpha=0.3)
    ax.set_title(stem_cats[sp-1])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if sp == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 6:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom="on")
        
for sp in range(1,6):
    ax=fig.add_subplot(6,3,sp+1+(sp-1)*2)
    ax.plot(women_degrees["Year"],women_degrees[lib_arts_cats[sp-1]],c=cb_dark_blue,label="Women",linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[lib_arts_cats[sp-1]],c=cb_orange, label="Men",linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255,171/255,171/255),alpha=0.3)
    ax.set_title(stem_cats[sp-1])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if sp == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom="on")
        
for sp in range(1,7):
    ax=fig.add_subplot(6,3,sp+2+(sp-1)*2)
    ax.plot(women_degrees["Year"],women_degrees[other_cats[sp-1]],c=cb_dark_blue,label="Women",linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[other_cats[sp-1]],c=cb_orange,label="Men",linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50,c=(171/255,171/255,171/255),alpha=0.3)
    ax.set_title(stem_cats[sp-1])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom="off")
    
    if sp == 1:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 6:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom="on")
    
plt.savefig("gender_degrees.png")
    
plt.show()
```
